```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import os
print matplotlib.__version__
```

```python
files = os.listdir('./Data/')
# print files
faces_paths = []
labels = []

for f in files:
    ims = os.listdir('./Data/' + f)
    # print f, ims
    for px in ims:
        faces_paths.append('./Data/' + f + '/' + px)
        labels.append(f)

print faces_paths
print labels
```

```python
imgs=[]

for fx in faces_paths:
    im=cv2.imread(fx)
    im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    imgs.append(im)
   
print len(imgs)
```

```python
# for ix in range(len(imgs)):
#     plt.figure(ix)
#     plt.imshow(imgs[ix])

# plt.show()
```

```python
all_faces=[]
plt.figure(0)
for mm in range(5):
    im=imgs[mm][100:500,550:850,:]
    print im.shape
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    print img.shape
    all_faces.append(img)
    plt.imshow(img,cmap='gray')
    plt.show()
    
for mm in range(5,9):
    im=imgs[mm][100:590,500:800,:]
    print im.shape
    img=cv2.resize(im,(100,100))
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    all_faces.append(img)
    print img.shape
    plt.imshow(img,cmap='gray')
    plt.show()    
    
# im=imgs[9][100:600,575:900,:]
# print im.shape
# img=cv2.resize(im,(100,100))
# img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
# all_faces.append(img)
# print img.shape
# plt.imshow(img,cmap='gray')
# plt.show()        
    
```
